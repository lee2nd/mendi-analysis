```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("mendi_record.csv")
df
```

```python
df_3m = df[df["Duration (minutes)"]==3]
df_3m["Date"] = df_3m["Year"].apply(str) + "-" + df_3m["Date"]
df_3m["Date"] = df_3m["Date"].str.replace("/","-")
df_3m['Date'] = pd.to_datetime(df_3m['Date'])
df_3m = df_3m.drop(columns=["Year","Duration (minutes)"])
df_3m
```

```python
# è¨­ç½®åœ–å½¢å¤§å°
plt.figure(figsize=(10, 6))

# ç¹ªè£½æŠ˜ç·šåœ–
plt.plot(df_3m['Date'], df_3m['Neural Activity (+%)'], label='Neural Activity (+%)')
plt.plot(df_3m['Date'], df_3m['Control (seconds)'], label='Control (seconds)')
plt.plot(df_3m['Date'], df_3m['Score'], label='Score')

# ç¹ªè£½å¹³å‡ç·š
plt.axhline(df_3m['Score'].mean(), color='black', linestyle='--', label='Average Score')

# è¨­ç½®æ¨™é¡Œå’Œè»¸æ¨™ç±¤
plt.title('Score Mean: ' + str(df_3m['Score'].mean()))
plt.xlabel('Date')
plt.ylabel('Value')

# æ·»åŠ åœ–ä¾‹
plt.legend()
```

```python
# è¨­ç½®åœ–å½¢å¤§å°
plt.figure(figsize=(10, 6))

# ç¹ªè£½æŠ˜ç·šåœ–
plt.plot(df_3m['Date'], df_3m['Resilience (p)'], label='Resilience (p)')

# ç¹ªè£½å¹³å‡ç·š
plt.axhline(df_3m['Resilience (p)'].mean(), color='black', linestyle='--', label='Average Resilience (p)')

# è¨­ç½®æ¨™é¡Œå’Œè»¸æ¨™ç±¤
plt.title('Resilience (p): ' + str(df_3m['Resilience (p)'].mean()))
plt.xlabel('Date')
plt.ylabel('Value')

# æ·»åŠ åœ–ä¾‹
plt.legend()
```
